# Druckzeitkonverter

* konvertiere Tage:Stunden:Minuten in total Minutes
* ***Eingabeform: 00:00:00***
* ***Ausgabeform: XXXX Minuten***

```python
a = input()
print('Eingabe: ' + str(a))

def time_split(a):

    string_list = a.split(':')   # splits the time 
    days = int(string_list[0])
    hours = int(string_list[1])
    minutes = int(string_list[2])

    days = days * 24
    hours = (hours + days)*60
    minutes = hours + minutes
    return minutes

print('Der Druck dauert: ' + str(time_split(a)) + ' Minuten.')
```
