# 8-ball for making decisions

+ random choosing from a list of possibilities
+ giving an answer

```python
import random
import tkinter as tk

print('What is your question?')
print()
print('Please formulate it in a yes/no way!')
print()

question = input()
answer = ['yes', 'no', 'maybe', 'likely yes', 'likely no']
ind = random.randint(0, len(answer)-1)
answer2 = answer[ind]
print()
print()
print('Answer:' + answer2)
print()
```

```python



root = tk.Tk()
label = tk.Label(root, text = 'What is your question?')
label.pack()
label2 = tk.Label(root, text = 'Please formulate it in a yes/no way!')
label2.pack()
bQuit = tk.Button(root, text = 'Quit', command=root.destroy)
bQuit.pack()
bRun = tk.Button(root, text = 'Give me all the answers!', command = AnswerLabel())
bRun.pack()

root.mainloop()

def AnswerLabel():
    aL = answer = tk.Label(root, text = answer2)
    aL.pack()
```
