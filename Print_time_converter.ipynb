```python
import tkinter as tk
def time_split(a):               # funtion for time convertion (splits days, hours, minutes in total minutes)
    a = e1.get()                 # input in dd:hh:minmin 
    days, hours, minutes = a.split(':')[:3]  # string split into a list

    days = int(days) * 24             # calculation of hours from days
    hours = (int(hours) + days)*60    # minutest from summ of days
    minutes = hours + int(minutes)
    resultLabel.config(text=minutes) # putting the result of the calculation into the result Label for output
    e1.delete(0,END) 
    #return minutes
    
    
master = tk.Tk() # initialisation of the GUI
master.title('Print time converter') #Title
tk.Label(master, text = 'Input for print time calculation: dd:hh:mm :').pack()
e1 = tk.Entry(master) # manual user input
e1.bind("<Return>",time_split) # 'Enter' returns the value to the calculation function
e1.pack() # everything packed & ready to go
outputLabel = tk.Label(master, text = 'Total minutes needed for the print:').pack() # generating output label
resultLabel = tk.Label(master, text = "") # empty output label
resultLabel.pack(fill=tk.X) # filling the output label with the result of the function
master.mainloop() #starting
```
