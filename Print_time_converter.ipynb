```python
import tkinter as tk
from tkinter import *
from math import *
```

```python
def time_split(a):               # funtion for time convertion (splits days, hours, minutes in total minutes)
    a = e1.get()                 # input in dd:hh:minmin 
    string_list = a.split(':')   # string split into a list
    days = int(string_list[0])
    hours = int(string_list[1])
    minutes = int(string_list[2])

    days = days * 24             # calculation of hours from days
    hours = (hours + days)*60    # minutest from summ of days
    minutes = hours + minutes
    resultLabel.config(text=minutes) # putting the result of the calculation into the result Label for output
    e1.delete(0,END) 
    #return minutes
    
    
master = Tk() # initialisation of the GUI
master.title('Print time converter') #Title
tk.Label(master, text = 'Input for print time calculation: dd:hh:mm :').pack()
e1 = tk.Entry(master) # manual user input
e1.bind("<Return>",time_split) # 'Enter' returns the value to the calculation function
e1.pack() # everything packed & ready to go
outputLabel = Label(master, text = 'Total minutes needed for the print:').pack() # generating output label
resultLabel = Label(master, text = "") # empty output label
resultLabel.pack(fill=X) # filling the output label with the result of the function
master.mainloop() #starting
```
